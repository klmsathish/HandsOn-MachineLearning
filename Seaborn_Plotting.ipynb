## Distribution plots
1.Distplot

2.Jointplot

3.pairplot



```python
sb.distplot(tips["total_bill"])
```

```python
#kde refers density estimation,bins refers size of tabs
sb.distplot(tips["total_bill"],kde=False,bins=50)
```

```python
sb.jointplot(x= "total_bill",y="tip",data = tips)
```

```python
#pairplotiing required columns
sb.pairplot(tips,hue = "sex")
```

```python
#pairplotiing whole data
sb.pairplot(tips,hue="smoker",palette = "coolwarm")
```

## Categorical Plots
1.Barplot

2.Countplot

3.Boxplot

4.ViolinPlot





```python
#barplot(x="",y = "",data = variable saved dataframe)
sb.swarmplot( x = "total_bill",data=tips)
```

```python
#barplot wwith extra parameters(x = "",y= "",data = datframe,estimator = some basic operations)
sb.barplot(x = "sex",y="total_bill",data=tips,estimator = np.std)
```

```python
#countplot(counting numbers of each data)
sb.countplot(x="sex",data=tips)
```

```python

sb.countplot(x="Event",data= df,alpha=0.5)
```

```python
#violin(x = ,y= hue = 3rd categorization i n a single graph ,)
sb.violinplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
#boxplotting(x = " ",y = " ", data = datset name)
sb.boxplot(x="day",y="total_bill",data = tips)
```
