```python
#confusion matrix to visualize matrices
from sklearn.metrics import confusion_matrix
#accuracy_score to visualize accuracy of our model
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
#original output
import numpy as np
actual = np.array([1.,0,0.0,1,0,0,1,0])
#predicted output
predicted = np.array([1,0,1.0,1,1.0,0,0,0])
print(type(actual), actual.shape, type(predicted))
```

```python
#fitting our actal and prediction to confusion matrix model
results= confusion_matrix(actual,predicted)
print("Confusion Matrix : ")
print(results)

#printing accuracy scores
print("Accuracy score : ",accuracy_score(actual,predicted))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyp
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn import metrics
%matplotlib inline
```

```python
#for locking test and train separatiion.similar to "Seed"
random_state = 0
```

```python
data = pd.read_csv("Weather_new.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
X = data['MinTemp'].values.reshape(-1,1)
y = data['MaxTemp'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#comparing Actual and Predicted
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
#Predicting errors
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#accuracy score of our model
print("Accuracy score : ",accuracy_score(y_test.round(), y_pred.round()))
```

```python
#np.ravel(y_train,0)
y_train.reshape(-1)
```

```python
#encoding values for logistic regression
lab_enc = preprocessing.LabelEncoder()
encoded = lab_enc.fit_transform(X_train)
encoded1 = lab_enc.fit_transform(y_train)
```

```python
#calling an instance of model
logregression = LogisticRegression()
logregression.fit(encoded.reshape(-1,1),encoded1)
```

```python
#predicting y value
y_pred = logregression.predict(X_test)
```

```python
#accuracy of same dataset withlogistic regression decreases
accuracy_score(y_test.round(), y_pred.round())
```

## Multipe categories

```python
#importing our data
dataset = pd.read_csv("winequality.csv")
```

```python
dataset.shape
```

```python
#decribe about mean,sts etc.
dataset.describe()
```

```python
#checking whether any column is null
dataset.isnull().any()
```

```python
#if any is null filling it with ffill method(ffill = forwardfill(copies next value to blank cells))
dataset = dataset.fillna(method='ffill')
```

```python
#dividinf labels
X = dataset[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates','alcohol']].values
y = dataset['quality'].values
```

```python
#splitting our data(test = 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#FIITING OUR MODEL
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
#CHECKING REGRESSOR AND COEFFICIENT
x = pd.DataFrame(X)
coeff_df = pd.DataFrame(regressor.coef_, x.columns,columns = [ "coeffieient"])
coeff_df
```

```python
#predicting y values
y_pred = regressor.predict(X_test)
```

```python
#comparing actual and predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
#finding our errors of our model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#Accuracy score of linear regression model
print("Accuracy score in Linear Reegression : ",accuracy_score(y_pred.round(),y_test))
```

```python
#fitting logistic regression model
logregressor = LogisticRegression()
logregressor.fit(X_train,y_train)
```

```python
#predicting for logistic regression
y_pred = logregressor.predict(X_test)
```

```python
#accuracy score hereaslo reduces for logistic regression model
print("Accuracy Score in Logistic Regression - ",accuracy_score(y_pred,y_test))
```
