```python
def convert_notavailable_nan(cell):
    if cell == "not available " or len(cell) == 0:
        return np.nan
    return cell
    
```

```python
df = pd.read_excel("weather.xlsx","weather",converters = {"Event":convert_notavailable_nan})
```

```python
df.loc[df["Event"].isnull()]
```
