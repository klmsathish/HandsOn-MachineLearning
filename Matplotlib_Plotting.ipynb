```python
#importing matplotlib
import matplotlib.pyplot as pyp
```

```python
%matplotlib inline 
```

```python
#importing numpy
import numpy as np
```

```python
#creating uniformly distributed values
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
#plooting a graph

pyp.plot(x,y,"red",marker = "*")
pyp.xlabel("sample")
pyp.ylabel("python")
pyp.title("data")
```

```python
#printing subplots(nrows,ncolumns,positions)
pyp.subplot(2,3,1)
pyp.plot(x,y,"r")
pyp.subplot(2,3,(5,6))
pyp.plot(x,y,"r")
pyp.subplot(2,3,3)
pyp.plot(x,y,"r")
```

```python
#printing multiple sublots
pyp.subplot(2,3,(1,2))
pyp.plot(x,y,"r")
pyp.subplot(2,3,3)
pyp.plot(x*2,y*2,"b")
pyp.subplot(2,3,4)
pyp.plot(x-1,y-1,"pink")
pyp.subplot(2,3,(5,6))
pyp.plot(x-1,y-1,"r")

```
