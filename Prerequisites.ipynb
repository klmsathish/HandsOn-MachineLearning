```python
#confusion matrix to visualize matrices
from sklearn.metrics import confusion_matrix
#accuracy_score to visualize accuracy of our model
from sklearn.metrics import accuracy_score
```

```python
#original output
actual = [1,0,2,1]
#predicted output
predicted = [1,0,8,3]
```

```python
#fitting our actal and prediction to confusion matrix model
results= confusion_matrix(actual,predicted)
print("Confusion Matrix : ")
print(results)
#printing accuracy scores
print("Accuracy score : ",accuracy_score(actual,predicted))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame({
    'x': [12, 20, 28, 18, 29, 33, 24, 45, 45, 52, 51, 52, 55, 53, 55, 61, 64, 69, 72,45,12,78,56,23,78,12,78,25,45],
    'y': [39, 36, 30, 52, 54, 46, 55, 59, 63, 70, 66, 63, 58, 23, 14, 8, 19, 7, 24,9,78,12,43,47,20,45,96,41,78]
})


np.random.seed(200)
k = 3
# centroids[i] = [x, y]
centroids = {
    i+1: [np.random.randint(0, 80), np.random.randint(0, 80)]
    for i in range(k)
}
    
fig = plt.figure(figsize=(5, 5))
plt.scatter(df['x'], df['y'], color='k')
colmap = {1: 'r', 2: 'g', 3: 'b'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlim(0, 80)
plt.ylim(0, 80)
plt.show()
```
